```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x ** 2
y
```

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Trial')
```

### using oop

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.4])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_xlabel('xlabel')
axes2.set_ylabel('ylabel')
```

```python
fig,axes = plt.subplots()
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1)
plt.tight_layout()
axes[0].plot(x,y)
```

```python
type(fig)
```

```python
fig = plt.figure(figsize=(3,3))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,ax = plt.subplots(nrows=1,ncols=3,figsize=(8,3))
ax[0].plot(x,y)
ax[1].plot(y,x)
```

```python
fig.savefig('matplotlib_practice.png')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,label='first',color="green",linewidth=3,linestyle='--')
axes.legend()
```
